```python
import os
import pandas as pd
from IPython.display import display

pd.options.display.max_columns = None
emp_data = pd.read_csv('../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.tsv.txt',sep='\t')
```

```python
print(emp_data.describe(include='all'))
```

```python
print(emp_data.info())
```

```python
# Confirm employee numbers are unique identifiers and set this column as the index

print(len(emp_data['EmployeeNumber'].unique()))
emp_data = emp_data.set_index('EmployeeNumber').sort_index()
print(emp_data.head())
```

```python
# Convert certain fields from numerical to categorical
categoricals = ['Education','JobLevel']
emp_data[categoricals] = emp_data[categoricals].astype('category')
print(emp_data.info())
```

```python
# Remove columns that have no variation in values
drops = ['EmployeeCount','StandardHours','Over18']
emp_data.drop(drops,axis=1,inplace=True)
print(emp_data.info())
```

```python
#save intermediate data file

emp_data.to_csv('../data/interim/emp_data_cleaned.tsv',sep='\t')
```
